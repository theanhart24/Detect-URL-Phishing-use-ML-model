```python
def getDomain(url):  
  domain = urlparse(url).netloc
  if re.match(r"^www.",domain):
	       domain = domain.replace("www.","")
  return domain
df['Domain']=df['url'].apply(lambda i: getDomain(i))
# 2.Checks for IP address in URL (Have_IP)
def havingIP(url):
  try:
    ipaddress.ip_address(url)
    ip = 1
  except:
    ip = 0
  return ip
df['havingIP']=df['url'].apply(lambda i: havingIP(i))
# 3.Checks the presence of @ in URL (Have_At)
def haveAtSign(url):
  if "@" in url:
    at = 1    
  else:
    at = 0    
  return at
df['haveAtSign']=df['url'].apply(lambda i: haveAtSign(i))
#f1 : url length(u) :

def urlLenght(url):
    length=len(url)
    return length
#tarchr(url) - phân bố các ký tự đặc biệt trong url tarchar(url)= countchar(url)/len(url)
df['urlLenght'] = df['url'].apply(lambda i: urlLenght(i))
def tarChar(url):
    special_characters = "!@#$%^&*()-_+=[]{}|:;\"'<>,.?/"
    count = 0 
    for char in url:
        if char in special_characters:
            count+=1
    return float(count)/len(url)
df['tarChar']=df['url'].apply(lambda i: tarChar(i))
#hashKeywords(url) : trả về giá trị 1 nếu tồn tại các từ khóa , else return 0
def hasKeywords(url): 
    hashkeyword = [
        "login","password","verify","account","secure","update","banking",
        "paypal","confirm","lostPassword","garrett_backstrom_wikipedia"
        #thảo luận thêm các từ khóa sau này để hoàn thiện chức năng của hàm ==> nên tạo 1 file chứa các keywords rồi import vào cho gọn 
    ]
    #tiến hành tạo 1 tập chứa các giá trị băm của các từ khóa
    hashed_Keywords=[]
    for keyword in hashkeyword:
        hashed_Keywords.append(hashlib.md5(keyword.encode()).hexdigest())
    #Tạo mảng chứa các hash keyword của url được dùng để đối chiếu với hashedKeywords
    hashed_url=[]
    for key in url.split('/'):#tiến hành tác các từ khóa ra theo ký tự / 
#         print(key)
        hashed_url.append(hashlib.md5(key.encode()).hexdigest())
    for key_final in hashed_url:
        if key_final in hashed_Keywords:
            return 1
    return 0
df['hasKeyWords']=df['url'].apply(lambda i: hasKeywords(i))
#f4: hasSpechar(url): trả về giái trị 1 nếu tồn tại các từ khóa , gược lại trả về giá trị là 0
#tuong tự như hàm trên nên cần bàn lại có thể copy paste
#f5: hashSpeKW(url): trả về giá trị 1 nếu tồn tại từ nhạy cảm tương tự luon được cái không cần mã hóa
def hasSpeChar(url):
    match = re.search('bit\.ly|goo\.gl|shorte\.st|go2l\.ink|x\.co|ow\.ly|t\.co|tinyurl|tr\.im|is\.gd|cli\.gs|'
                      'yfrog\.com|migre\.me|ff\.im|tiny\.cc|url4\.eu|twit\.ac|su\.pr|twurl\.nl|snipurl\.com|'
                      'short\.to|BudURL\.com|ping\.fm|post\.ly|Just\.as|bkite\.com|snipr\.com|fic\.kr|loopt\.us|'
                      'doiop\.com|short\.ie|kl\.am|wp\.me|rubyurl\.com|om\.ly|to\.ly|bit\.do|t\.co|lnkd\.in|'
                      'db\.tt|qr\.ae|adf\.ly|goo\.gl|bitly\.com|cur\.lv|tinyurl\.com|ow\.ly|bit\.ly|ity\.im|'
                      'q\.gs|is\.gd|po\.st|bc\.vc|twitthis\.com|u\.to|j\.mp|buzurl\.com|cutt\.us|u\.bb|yourls\.org|'
                      'x\.co|prettylinkpro\.com|scrnch\.me|filoops\.info|vzturl\.com|qr\.net|1url\.com|tweez\.me|v\.gd|'
                      'tr\.im|link\.zip\.net',
                      url)
    if match: return 1 
    else: return 0
df['hasSpeChar']=df['url'].apply(lambda i: hasSpeChar(i))
#f7 : tahex(url): phân bố ký tự hexa trong URL 
#tahex(url)= counthe(url)/len(u) trong đó counthe là số ký tự hexa
def tahex(url):
    count_hex=sum(1 for char in url if char in "0123456789abcdefABCDEF")
    length_url=len(url)
    if length_url!=0:
        ratio=float(count_hex)/length_url
    else:
        ratio=0
    return ratio

df['tarhex']=df['url'].apply(lambda i: tahex(i))
#Phân bố chữ số có trong URL 
#tardigit(url) = count_digit(url)/len(url)
def tadigit(url):
    count_number=sum(1 for number in url if number in "0123456789")
    if(len(url)!=0):
        tadigit=float(count_number)/len(url)
    else:
        tadigit=0
    return tadigit
df['Tadigit']=df['url'].apply(lambda i: tadigit(i))

def numdots(url):
    count_dot=url.count('.')# hàm count trong python được dùng để đếm số lần xuất hiện của 1 ký tự hoặc 1 chuỗi trong 1 string hoặc objects.
    return count_dot
df['numdots']=df['url'].apply(lambda i: numdots(i))
def taslash(url):
    count_taslash=url.count('/')
    if count_taslash !=0:
        talash_final=float(count_taslash)/len(url)
    else:
        talash_final=0
    return talash_final
df['taslash']=df['url'].apply(lambda i: taslash(i))
def countcase(url):
    count_case=sum(1 for upper_case in url if upper_case in "QWERTYUIOPASDFGHJKLZXCVBNM")
    return count_case

df['countcase']=df['url'].apply(lambda i: countcase(i))
def numvo(url):
    count_vo=sum(1 for dipthong in url if dipthong in "aiuoeAIUOE")
    numvo=float(count_vo)/len(url)
    return numvo
df['numvo']=df['url'].apply(lambda i: numvo(i))

#kiểm tra số subdomain xuất hiện trong url 
def detect_subdomain_using_regex(url):
    subdomains = re.findall(r'(?<=\/\/)([^\/]+)', url)
    return len(subdomains)
#########phần của thiện #################
def haspro(url):
    #tim kien cac chuoi http,https,www trong url
    http_found = "http" in url.lower()
    https_found= "https" in url.lower()
    www_found = "www" in url.lower()
    if http_found or https_found or www_found:
        return 1
    else : return 0
df['haspro']=df['url'].apply(lambda i: haspro(i))
def hasExe(url):
    #tim kien cac chuoi http,https,www trong url
    exe_found = ".exe" in url.lower()
    if exe_found:
        return 1
    else : return 0
def has_port(url):
  # Sử dụng `urlparse` để phân tích URL.
  parsed_url = urlparse(url)

  # Kiểm tra xem port có khác None hay không.
  return 1 if parsed_url.port else 0
df['has_port']=df['url'].apply(lambda i: has_port(i))
def backslash(url):
    if "\\" in url:
        return 1
    else:
        return 0
df['backslash']=df['url'].apply(lambda i: backslash(i))
# def redirect(url):
#     try:
#         response = requests.get(url)
#         if response.status_code in [301, 302]:
#             return 1
#         else:
#             return 0
#     except requests.exceptions.RequestException as e:
#         print("Error:", e)
#         return 0
# df['redirect']=df['url'].apply(lambda i: redirect(i))
def hasref(url):
    phishing_keywords = ["ref=", "cdm=", "login=", "account=", "password=", "verify=", "secure="]  # Danh sách các từ khóa phishing
    for keyword in phishing_keywords:
        if keyword in url:
            return 1  # Nếu bất kỳ từ khóa phishing nào được tìm thấy trong URL, trả về 1
    return 0
df['hashref']=df['url'].apply(lambda i: hasref(i))

# 5.Gives number of '/' in URL (URL_Depth)
def getDepth(url):
  s = urlparse(url).path.split('/')
  depth = 0
  for j in range(len(s)):
    if len(s[j]) != 0:
      depth = depth+1
  return depth
df['getDepth']=df['url'].apply(lambda i: getDepth(i))
# 6.Checking for redirection '//' in the url (Redirection)
def redirection(url):
  pos = url.rfind('//')
  if pos > 6:
    if pos > 7:
      return 1
    else:
      return 0
  else:
    return 0
df['redirection']=df['url'].apply(lambda i: redirection(i))
# 7.Existence of “HTTPS” Token in the Domain Part of the URL (https_Domain)
def httpDomain(url):
  domain = urlparse(url).netloc
  if 'https' in domain:
    return 1
  else:
    return 0
df['httpDomain']=df['url'].apply(lambda i: httpDomain(i))
#listing shortening services
shortening_services = r"bit\.ly|goo\.gl|shorte\.st|go2l\.ink|x\.co|ow\.ly|t\.co|tinyurl|tr\.im|is\.gd|cli\.gs|" \
                      r"yfrog\.com|migre\.me|ff\.im|tiny\.cc|url4\.eu|twit\.ac|su\.pr|twurl\.nl|snipurl\.com|" \
                      r"short\.to|BudURL\.com|ping\.fm|post\.ly|Just\.as|bkite\.com|snipr\.com|fic\.kr|loopt\.us|" \
                      r"doiop\.com|short\.ie|kl\.am|wp\.me|rubyurl\.com|om\.ly|to\.ly|bit\.do|t\.co|lnkd\.in|db\.tt|" \
                      r"qr\.ae|adf\.ly|goo\.gl|bitly\.com|cur\.lv|tinyurl\.com|ow\.ly|bit\.ly|ity\.im|q\.gs|is\.gd|" \
                      r"po\.st|bc\.vc|twitthis\.com|u\.to|j\.mp|buzurl\.com|cutt\.us|u\.bb|yourls\.org|x\.co|" \
                      r"prettylinkpro\.com|scrnch\.me|filoops\.info|vzturl\.com|qr\.net|1url\.com|tweez\.me|v\.gd|" \
                      r"tr\.im|link\.zip\.net"
# 8. Checking for Shortening Services in URL (Tiny_URL)
def tinyURL(url):
    match=re.search(shortening_services,url)
    if match:
        return 1
    else:
        return 0
df['tinyURL']=df['url'].apply(lambda i: tinyURL(i))
# 9.Checking for Prefix or Suffix Separated by (-) in the Domain (Prefix/Suffix)
def prefixSuffix(url):
    if '-' in urlparse(url).netloc:
        return 1            # phishing
    else:
        return 0            # legitimate
df['prefixSuffix']=df['url'].apply(lambda i: prefixSuffix(i))
# 12.Web traffic (Web_Traffic)
# def web_traffic(url):
#   try:
#     #Filling the whitespaces in the URL if any
#     url = urllib.parse.quote(url)
#     rank = BeautifulSoup(urllib.request.urlopen("http://data.alexa.com/data?cli=10&dat=s&url=" + url).read(), "xml").find(
#         "REACH")['RANK']
#     rank = int(rank)
#   except TypeError:
#         return 1
#   if rank <100000:
#     return 1
#   else:
#     return 0
# df['web_traffic']=df['url'].apply(lambda i: web_traffic(i))
# 13.Survival time of domain: The difference between termination time and creation time (Domain_Age)  
# def domainAge(domain_name):
#   creation_date = domain_name.creation_date
#   expiration_date = domain_name.expiration_date
#   if (isinstance(creation_date,str) or isinstance(expiration_date,str)):
#     try:
#       creation_date = datetime.strptime(creation_date,'%Y-%m-%d')
#       expiration_date = datetime.strptime(expiration_date,"%Y-%m-%d")
#     except:
#       return 1
#   if ((expiration_date is None) or (creation_date is None)):
#       return 1
#   elif ((type(expiration_date) is list) or (type(creation_date) is list)):
#       return 1
#   else:
#     ageofdomain = abs((expiration_date - creation_date).days)
#     if ((ageofdomain/30) < 6):
#       age = 1
#     else:
#       age = 0
#   return age
# df['domainAge']=df['url'].apply(lambda i: domainAge(i))

# # 17.Checks the status of the right click attribute (Right_Click)
# def rightClick(response):
#   if response == "":
#     return 1
#   else:
#     if re.findall(r"event.button ?== ?2", response.text):
#       return 0
#     else:
#       return 1
# df['rightClick']=df['url'].apply(lambda i: rightClick(i))
```
